```python
import sys

import requests
from bs4 import BeautifulSoup
import re
import unicodedata
import pandas as pd
import json
```

```python
static_url = "https://www.skysports.com/premier-league-results/2019-20"
```

```python
response =  requests.get(static_url) 
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup.title
```

```python
html_div = soup.find_all("div", {"class":"fixres__item"})
#html_div2 = soup.find("script", type="text/show-more")
count = 0
matchResult=[]
```

```python

for res in html_div:
    res = res.text
    res = res.split()
    if res[1].isnumeric() is False:
        res[0:2] = [''.join(res[0 : 2])]
        if res[1].isnumeric() is False:
            res[0:2] = [''.join(res[0 : 2])]
            if res[1].isnumeric() is False:
                res[0:2] = [''.join(res[0 : 2])]
    res.pop(3)
    if res[3] == 'Watford':
        res = res[0:4]
    else:
        res = res[:-1]
    if len(res)==5:
        res[3:5] = [''.join(res[3 : 5])]
    if len(res)==6:
        res[3:6] = [''.join(res[3 : 6])]
    if len(res)==7:
        res[3:7] = [''.join(res[3 : 7])]
    res[1]=int(res[1])
    res[2]=int(res[2])
    res[1],res[3]=res[3],res[1]
    res[2],res[3]=res[3],res[2]
    print(res)
    matchResult.insert(0,res)
    count+=1
```

```python
from IPython.display import HTML
import base64, io

# Download as CSV: data frame, optional title and filename
def create_download_link_csv(df, title = "Download CSV file", filename = "data.csv"):  
    # generate in-memory CSV, then base64-encode it
    csv = df.to_csv(index=False)
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)
```

```python
create_download_link_csv(df, title = "Download CSV file", filename = "soccerResults.csv")
```

```python
# Download as Excel: data frame, optional title and filename
def create_download_link_excel(df, title = "Download Excel file", filename = "data.xlsx"):  
    # to_excel() does not work to string buffer directly
    output = io.BytesIO()
    # Use the BytesIO object as the filehandle
    writer = pd.ExcelWriter(output, engine='xlsxwriter')
    # Write the data frame to the BytesIO object and save it
    df.to_excel(writer, sheet_name='Sheet1')
    writer.save()
    excel_data = output.getvalue()
    b64 = base64.b64encode(excel_data)
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/xml;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)
```

```python
create_download_link_excel(df, title = "Download EXCEL file", filename = "soccerResults.xlsx")
```
